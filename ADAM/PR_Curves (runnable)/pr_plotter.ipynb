```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
names = ["spectral", "dtsh", "dpsh", "dsh", "subic", "triplet"]
bits = 32
dataset = "Nus_Wide"
```

```python
def pr_data(name, bits, dataset):
    df = pd.read_csv(f"/Users/kristiansjorslevnielsen/Documents/DVP7/DeepHashing/Pointwise/imagenet_tests/pr_curves/{name}_{dataset}_{bits}_pr_curve.csv")
    return df
```

```python

# Create the plot
plt.figure(figsize=(8, 6))  # Set figure size

# Plot multiple lines
for name in names:
    try:
        df = pr_data(name, bits, dataset)[1:]
        plt.plot(df["recall"], df["precision"], label=name)
    except:
        continue

# Add labels and title
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall curve for each model on {dataset}\n with {bits} bit hash codes')
plt.xlim(0,1)
plt.ylim(0,1)

# Add legend
plt.legend()

# Show the plot
plt.grid(True)  # Optional: add a grid for better readability
plt.show()
```

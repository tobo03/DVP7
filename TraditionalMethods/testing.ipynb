```python
def lshProjt(data, randVectors):
    data = data.flatten()
    return [int(np.dot(data, randVector) > 0) for randVector in randVectors]
```

```python
n_hash = 12
data_len = len(oneObs.flatten())
randVectors = [np.random.randn(data_len) for i in range(n_hash)]

lshProjt(oneObs, randVectors)
```

```python
hog_feat = hog(oneObs)

n_hash = 12
data_len = len(hog_feat)
randVectors = [np.random.randn(data_len) for i in range(n_hash)]

lshProjt(hog_feat, randVectors)
```

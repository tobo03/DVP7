# LSH (projection)

```python
oneObs =  x_train[0]
oneObs
```

```python
def lshProjt(data, randVectors):
    data = data.flatten()
    return [int(np.dot(data, randVector) > 0) for randVector in randVectors]
```

```python
n_hash = 12
data_len = len(oneObs.flatten())
randVectors = [np.random.randn(data_len) for i in range(n_hash)]

lshProjt(oneObs, randVectors)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

# Step 1: Perform K-Means Clustering
def kmeans_clustering(hash_codes, n_clusters=10, n_init=10, random_state=42):
    kmeans = KMeans(n_clusters=n_clusters, n_init=n_init, random_state=random_state)
    clusters = kmeans.fit_predict(hash_codes)
    return clusters

# Step 2: Build the label-cluster matrix
def build_label_cluster_matrix(clusters, true_labels, n_clusters=10):
    # Initialize a matrix to count label-cluster relationships  
    label_cluster_matrix = np.zeros((10, n_clusters), dtype=int)
    
    for label, cluster in zip(true_labels, clusters):
        label_cluster_matrix[label][cluster] += 1
    
    return label_cluster_matrix

# Step 3: Visualize the result as a heatmap
def plot_heatmap(label_cluster_matrix):
    plt.figure(figsize=(10, 6))
    sns.heatmap(label_cluster_matrix, annot=True, fmt="d", cmap="Blues", cbar=True)
    plt.title("Label Distribution Across Clusters")
    plt.xlabel("Clusters")
    plt.ylabel("True Labels")
    plt.show()


def easyHeatmap(hashCodes, labels):
    raw_clusters = kmeans_clustering(hashCodes, n_clusters=10)

    raw_label_cluster_matrix = build_label_cluster_matrix(raw_clusters, labels)

    plot_heatmap(raw_label_cluster_matrix)
```

```python
n_hash = 128
data_len = len(x_test[0].flatten())
randVectors = [np.random.randn(data_len) for i in range(n_hash)]

hashCodes = [ lshProjt(x, randVectors) for x in x_test]

easyHeatmap(hashCodes, y_test)
```

```python
n_hash = 128
data_len = len(hog(x_test[0]).flatten())
randVectors = [np.random.randn(data_len) for i in range(n_hash)]

hashCodes = [ lshProjt(hog(x), randVectors) for x in x_test]

easyHeatmap(hashCodes, y_test)
```

```python
n_hash = 128
data_len = len(x_test[0].flatten())

hashCodes = [ (x.flatten() > 0).astype(int) for x in x_test]

easyHeatmap(hashCodes, y_test)
```

```python
def evenSampling(data, n_hash):
    data = data.flatten()

    spaceing = len(data) / n_hash
    
    curr_i = 0

    output = []
    while curr_i < len(data):
        output.append(data[int(curr_i)])

        curr_i += spaceing


    return output

len( evenSampling(x_test[0], n_hash) )
```

```python
n_hash = 128
data_len = len(x_test[0].flatten())

hashCodes = [ evenSampling( (x.flatten() > 0).astype(int), n_hash) for x in x_test]

easyHeatmap(hashCodes, y_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Test\Desktop\p7\Spectral\results\spectral_results.csv')
```

```python
for bit in df["Bits"].unique():
    df_sub = df[df["Bits"]==bit]
    plt.plot(df_sub["Neighbours"], df_sub["MAP_10000"], label=bit)
plt.legend(title = "Bits")
plt.xlabel("Number of neighbours")
plt.ylabel("MAP")
```

```python
for bit in df["Bits"].unique():
    df_sub = df[df["Bits"]==bit]
    plt.plot(df_sub["Neighbours"], df_sub["P@100"], label=bit)
plt.legend(title = "Bits")
plt.xlabel("Number of neighbours")
plt.ylabel("P@100")
```
